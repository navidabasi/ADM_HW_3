# 2. Search Engine


## The process of making search engine using Natural Language ToolKit is described as below

    1)Tokenization
        Spliting text into pieces(Tokens), remove punctuation
    
    2)STOPWORD Removal
        Removes commonly used swords (Such as 'the') which are not relavant to analysis
        
    3)Stemming and Lemmatization:
        Reduces words to base form to be analyzed as a single item
    
    4)POS Tagging:
        Tags words to be part of speech (Such as verb, noun) based on definition and context
        
    5)Informational Retrieval:
        Extract relevant information from source

### Informational Retrieval

#### Making the idx

```python
count=Counter(reduce(lambda x, y: x + y, data.tokenized))
```

```python
word_dict = {} #dictionary with word as key and id as value
item_id = 1
for word in count.keys():
    word_dict[word] = item_id
    item_id += 1
```

```python
inverted_idx = {}
for word, item_id in word_dict.items():
    inverted_idx[item_id] = list(data[data.tokenized.apply(lambda row: word in row)].index)
```

```python
snowstem = SnowballStemmer('english')
def stem_text(txt):
    return [snowstem.stem(word) for word in txt.split(' ')]
```

```python
def search(query_str):
    data_conj=pd.DataFrame(columns=['placeName', 'placeDesc', 'placeURL'])
    results = [set(inverted_idx[item]) for item in [word_dict[word] for word in stem_text(query_str)]]
    conjuctive=set(reduce(lambda x,y: x.intersection(y), results))
    for doc in conjuctive:
        new_row = {'placeName': data.loc[doc]['placeName'], 'placeDesc': data.loc[doc]['desc'], 'placeURL': data.loc[doc]['url']}
        #append row to the dataframe
        
        data_conj = data_conj.append(new_row, ignore_index=True)
    data_conj=data_conj.dropna(subset=['placeName'])
    return data_conj
```

```python
#tf t,d = numero di parole /lunghezza campo
#idf t = log(N/df)  N is the total number of documents in the collection, and df is the number of documents a term occurs in. 
#w t,d = tf * idf The TF-IDF is the dot product of the TF and IDF, so calculated as:
inverted_idx2 = {}
term_idf = {}

for word, item_id in word_dict.items():
    result = []
    docList = inverted_idx[item_id]
    idf = math.log(len(data)/len(docList))
    for doc in docList:
        tf = (data.loc[doc].desc.count(word))/len(data.loc[doc].desc)
        tfidf = tf * idf
        result.append((doc, tfidf))
    term_idf[item_id] = idf
    inverted_idx2[item_id] = result
```

Calculating similarity between descriptions with Cosin Similarity

```python
def cosine(query, tokenized):
    query = query.strip().split()
    
    query = word_tokenize(query)
    l1 = []; l2 = []
    sw = stopwords.words('english') 

    query = {w for w in query if not w in sw} 
    
    vectorized = query.union(tokenized) 
    for w in vectorized:
        if w in query: l1.append(1) # create a vector
        else: l1.append(0)
        if w in tokenized: l2.append(1)
        else: l2.append(0)
    c = 0

    # cosine formula 
    for i in range(len(vectorized)):
            c+= l1[i]*l2[i]
    cosine = c / float((sum(l1)*sum(l2))**0.5)
    
    return cosine
```

```python
def search_2(query_str):
    results = [set(inverted_idx[item]) for item in [word_dict[word] for word in stem_text(query_str)]]
    conjuctive=set(reduce(lambda x,y: x.intersection(y), results))
    data_conj = data.loc[conjuctive][['placeName', 'placeDesc', 'placeURL']]
    data_conj["Similarity"] = [cosine(query_str, doc) for doc in conjuctive]
    data_conj = data_conj.sort_values(by="Similarity", ascending=False)
    return data_conj
```
